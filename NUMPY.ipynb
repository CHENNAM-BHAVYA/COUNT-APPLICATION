```python
import numpy as np
marks=np.array([78,58,18,96,100])
print("Average:",np.mean(marks))
print("Highest:",np.max(marks))
print("Lowest:",np.min(marks))
print("Sum:",np.sum(marks))
top=marks[marks>80]
print(top)
```
